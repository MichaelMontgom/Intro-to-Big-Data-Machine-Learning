```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats

try:

    # read in the Dataset (CSV) using pandas.
    # https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html
    # original dataset is found at https://www.kaggle.com/sohier/calcofi. Columnes F and G were used for Temperature (Celsius) and Salinity (grams of salt per kilograms of water (g/kg))

    data = pd.read_csv('data_columns.csv',low_memory=False)
    

    
    
    
    # slice the two rows
    temp = data.iloc[:,0]
    salt = data.iloc[:,1]
    
    # cast Panda series to Numpy arrays
    temp = temp.to_numpy()
    salt = salt.to_numpy()

    
    plt.hist(salt)
    plt.title('Distribution of Salinity variables')
    plt.savefig('Hist_Salinity.png', bbox_inches='tight')

    plt.hist(temp)
    plt.title('Distribution of Temperature variables')
    plt.savefig('Hist_temp.png', bbox_inches='tight')



    
#     # find the mean of the independent and dependent variables, make them float64s
#     temp_mean = np.nanmean(temp, dtype=np.float64)
#     salt_mean = np.nanmean(salt, dtype=np.float64)
    
#     b_top = 0
#     b_bot = 0
#     for i in range(len(temp)):
#         b_top = b_top + np.nan_to_num(((temp[i]-temp_mean)*(salt[i]-salt_mean)))
#         b_bot = b_bot + np.nan_to_num((temp[i]-temp_mean)**2)
#     b = b_top/b_bot
    
#     a = salt_mean - b*temp_mean


#     https://stackoverflow.com/questions/13643363/linear-regression-of-arrays-containing-nans-in-python-numpy
#     this is used to filter out all the non-values
    mask = ~np.isnan(temp) & ~np.isnan(salt)
    
#     https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.linregress.html
    slope, intercept, r_value, p_value, std_err = stats.linregress(temp[mask],salt[mask])
    
except Exception as e:
    print(e)
    
```

```python
try:
    f = lambda b0, b1, x: b0 + b1*x
    
    domain = np.linspace(1,np.nanmax(temp),1000)
    values = f(intercept,slope,domain)
    
    
    print(slope,intercept)
    

    
    
    
#     print(domain)
    fig, ax = plt.subplots(1,1)
#     ax.plot(temp, salt, 'rx')
    
#     data.plot(x='T_degC', y='Salnty', style='rx' )
#     plt.title('Temperature (C) vs Salinity (g/kg)')
#     plt.xlabel('Temperature (C)')
#     plt.ylabel('Salinity')
#     plt.savefig('Graph_of_total_points.png', bbox_inches='tight')
    ax.plot(temp,salt,'rx')
    ax.plot(domain, values,'g-')
    ax.set_title('Data with regression line')
    ax.set_xlabel('Temperature')
    ax.set_ylabel('Salinity')
    plt.savefig('regression.png')
#     print(values)

#     describe = data.describe().to_csv('description.csv')
    

    
except Exception as e:
    print(e)
```
